## Read Files

Reading two json files called "Jumping.json" and "Walking.json" into pandas dataframes
The both files were created via the 'Sensor Logger'-App:

- recording duration: ap. 10 Seconds
- one while walking and one while jumping around

```python
df_jump = pd.read_json("Jumping.json")
df_walk = pd.read_json("Walking.json")
```

## Preprocess some data

```python
# convert 'time' into date format

df_walk['time'] = pd.to_datetime(df_walk['time'])
df_jump['time'] = pd.to_datetime(df_jump['time'])

display(df_walk.head(2), df_jump.head(2))

# set time as index

df_walk = df_walk.set_index('time')
df_jump = df_jump.set_index('time')

display(df_walk.head(2), df_jump.head(2))
```

## Select and Plot some data

```python
# show the available sensors

df_jump['sensor'].unique()
```

```python
df_walk['sensor'].unique()
```

### Select and plot accelerometer data from df_walk

```python
# Select 'Accelerometer'
df_walk_acc = df_walk[df_walk['sensor'] == 'Accelerometer']
display(df_walk_acc.head(2))

df_jump_acc = df_jump[df_jump['sensor'] == 'Accelerometer']
display(df_jump_acc.head(2))
```

```python
# Select only relevant columns
df_walk_acc = df_walk_acc[['z', 'y', 'x']]
display(df_walk_acc.head(2))

df_jump_acc = df_jump_acc[['z', 'y', 'x']]
display(df_jump_acc.head(2))
```

```python
# Data recorded while walking
df_walk_acc.plot(figsize=(10,5))
```

```python
# Data recorded while jumping
df_jump_acc.plot(figsize=(10,5))
```
